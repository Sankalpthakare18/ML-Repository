```python
def sum_between(start, end):
    if start > end:
        return 0
    return start + sum_between(start + 1, end)
    
start_no = 1
end_no = 5
result = sum_between(start_no, end_no)
print(f"The sum of numbers between {start_no} and {end_no} is {result}")
```

```python
import chardet

# read the file in binary mode to detect encoding
with open(file_path, 'rb') as file_obj:
    raw_data = file_obj.read()
    result = chardet.detect(raw_data)
    encoding = result['encoding']

# open the file in read mode with detected encoding
with open(file_path, 'r', encoding=encoding) as file_obj:
    # read first character
    first_char = file_obj.read(1)

    if not first_char:
        print("File is empty")
    else:
        print("First character:", first_char)
```

```python
import requests
from bs4 import BeautifulSoup
import csv

def scrape_amazon(url):
  headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36'}
  response = requests.get(url, headers=headers)
  soup = BeautifulSoup(response.content, 'html.parser')

  try:
    price_element = soup.find('span', id="priceblock_ourprice").text.strip()
    title_element = soup.find('span', id="productTitle").text.strip()
    description_element = soup.find('div', id="feature-bullets").find_all('li')
    description = ', '.join([li.text.strip() for li in description_element])
  except AttributeError:
    return None, None, None

  return price_element, title_element, description

def save_to_csv(data, filename='amazon_products.csv'):
  with open(filename, 'a', newline='', encoding='utf-8') as csvfile:
    fieldnames = ['Price', 'Title', 'Description']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writerow({'Price': data[0], 'Title': data[1], 'Description': data[2]})

# Example usage
url = "https://www.amazon.com/Apple-iPhone-14-Pro-Max/dp/B0BL8F6V9F/ref=sr_1_1?crid=0232182034D9O8H7S90&keywords=iphone+14+pro+max&qid=1685291493&sprefix=iphone+14+pro+max%2Caps%2C%90%20%113&sr=8-1"
data = scrape_amazon(url)
if data:
  save_to_csv(data)

# Replace with your desired product URLs
urls = [
  "https://www.amazon.com/Apple-iPhone-14-Pro-Max/dp/B0BL8F6V9F/ref=sr_1_1?crid=0232182034D9O8H7S90&keywords=iphone+14+pro+max&qid=1685291493&sprefix=iphone+14+pro+max%2Caps%2C%90%113&sr=8-1",
  # Add more URLs here
]

for url in urls:
  data = scrape_amazon(url)
  if data:
    save_to_csv(data)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL of the Amazon page (ensure this is a valid mobile phone product page URL)
url = 'https://www.amazon.com/dp/B08N5WRWNW'  # Example URL for an iPhone product page

# Headers to mimic a real browser visit
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36',
}

# Send a request to fetch the page content
response = requests.get(url, headers=headers)

# Check if the request was successful
if response.status_code == 200:
    # Parse the page content
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Extract the data
    try:
        price = soup.find('span', {'id': 'priceblock_ourprice'}).text.strip()
    except AttributeError:
        price = 'Not Available'
    
    try:
        description = soup.find('div', {'id': 'productDescription'}).text.strip()
    except AttributeError:
        description = 'Not Available'

    try:
        amount = soup.find('span', {'id': 'acrCustomerReviewText'}).text.strip()
    except AttributeError:
        amount = 'Not Available'
    
    # Data dictionary
    data = {
        'Price': [price],
        'Amount': [amount],
        'Description': [description]
    }

    # Create a DataFrame
    df = pd.DataFrame(data)
    
    # Save the data to a CSV file
    df.to_csv('amazon_mobile_data.csv', index=False)
    
    print("Data has been successfully scraped and saved to 'amazon_mobile_data.csv'")
else:
    print(f"Failed to retrieve the page. Status code: {response.status_code}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

path = r"C:\Users\swamini\Downloads\titanic_original.csv"
titanic_data = pd.read_csv(path)

print(titanic_data.head())


gender_counts = titanic_data['sex'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(gender_counts, labels=gender_counts.index,  colors=['skyblue', 'lightpink'])
plt.title('Male/Female Proportion')
plt.show()

plt.figure(figsize=(10, 6))

# Plot males
male = titanic_data[titanic_data['sex'] == 'male']
plt.scatter(male['age'], male['fare'], color='blue', label='Male')

# Plot females
female = titanic_data[titanic_data['sex'] == 'female']
plt.scatter(female['age'], female['fare'], color='pink', label='Female')

plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot of Fare Paid vs Age')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

path = r"C:\Users\swamini\Downloads\diamonds.csv"
diamonds_data = pd.read_csv(path)

# Filter the dataset for clarity "SI2" and color "E"
filtered_data = diamonds_data[(diamonds_data['clarity'] == 'SI2') & (diamonds_data['color'] == 'E')]

plt.figure(figsize=(12, 8))

scatter = plt.scatter(filtered_data['carat'], filtered_data['price'], 
                      c=filtered_data['cut'].astype('category').cat.codes, 
                      cmap='viridis', edgecolors='w', s=100)


plt.xlabel('Carat')
plt.ylabel('Price')
plt.title('Relationship between Carat and Price for SI2 Clarity and E Color Diamonds')

legend1 = plt.legend(*scatter.legend_elements(), title="Cut")
plt.gca().add_artist(legend1)

plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
file_path = r"C:\Users\swamini\Downloads\planets.csv"
planets_data = pd.read_csv(file_path)

masses = planets_data['mass'].dropna()


print(masses.describe())

# Create a histogram
plt.figure(figsize=(12, 6))
plt.hist(masses, bins=50, color='skyblue', edgecolor='black')
plt.title('Distribution of Planet Masses')
plt.xlabel('Mass (in multiples of Jupiter\'s mass)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Create a box plot
plt.figure(figsize=(8, 6))
plt.boxplot(masses, vert=False, patch_artist=True, 
            boxprops=dict(facecolor='lightblue', color='black'),
            medianprops=dict(color='red'))
plt.title('Box Plot of Planet Masses')
plt.xlabel('Mass (in multiples of Jupiter\'s mass)')
plt.grid(True)
plt.show()
```
